```python
import pandas as pd;
import matplotlib.pyplot as plt;

ca = pd.read_csv('/workspaces/data_analytics/Week_02/GK/Cars_autoscout24.csv',sep=';', encoding='windows-1252');

ca['Price_new'] = ca['Price'].str.replace('[^\d]', '', regex=True);

# print(ca['Price_new'])

ca['Prices_numeric'] = pd.to_numeric(ca['Price_new'], errors='coerce')

plt.hist(ca['Prices_numeric'], bins=20)  # Adjust the number of bins as needed
plt.xlabel('Price (CHF)')
plt.ylabel('Frequency')
plt.title('Histogram of car prices')
plt.grid()
plt.show()
```
